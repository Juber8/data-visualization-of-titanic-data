```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_data.csv')
```

```python
df.head()
```

```python
g= sns.countplot(df['Survived'])
g.set_xticklabels(['not survived','survived'])
plt.show()
```

```python
df.Survived.value_counts()
```

```python
df.isna().sum()
```

```python
# so here cabin have 687 missing value out of 891
len(df)
```

```python
df.drop(columns='Cabin',inplace=True)
```

```python
df
```

```python
df[['Age']].isna().sum()
```

```python
df.fillna(df.Age.mean(),inplace=True)
```

```python
df['Age'].isna().sum()
```

```python
# here we in age columns we have 177 missing value we fill the values with the mean 
```

```python
plt.figure(figsize=(10,10))
plt.pie(df.Survived.value_counts(),autopct='%.2f%%',pctdistance=0.8)
plt.xlabel('survived')
plt.ylabel('not survived')
plt.show()
```
